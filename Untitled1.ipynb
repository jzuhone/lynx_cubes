```python
%matplotlib inline
import matplotlib.pyplot as plt
from astropy.io import fits
from matplotlib.colors import LogNorm
import numpy as np
from astropy.convolution import Gaussian2DKernel, convolve
import h5py
```

```python
fc = fits.open("cgols_cts_90deg.fits")
fv = fits.open("cgols_vel_90deg.fits")
```

```python
plt.figure(figsize=(10,10))
plt.imshow(fc[0].data, origin='lower', norm=LogNorm())
```

```python
plt.figure(figsize=(10,10))
plt.imshow(fv[0].data, origin='lower')
```

```python
def red(vel):
    red = np.zeros_like(vel)
    red[vel >= 0.0] = vel[vel >= 0.0]/200.0
    return red
```

```python
def green(vel):
    green = np.zeros_like(vel)
    green[vel < 0.0] = vel[vel < 0.0]/200.0+1.0
    green[vel >= 0.0] = -vel[vel >= 0.0]/200.0+1.0
    return green
```

```python
def blue(vel):
    blue = np.zeros_like(vel)
    blue[vel < 0.0] = -vel[vel < 0.0]/200.0
    return blue
```

```python
kernel = Gaussian2DKernel(x_stddev=1.0)
```

```python
redd = convolve(red(fv[0].data), kernel)
blued = convolve(blue(fv[0].data), kernel)
greend = convolve(green(fv[0].data), kernel)
```

```python
alpha = np.log10(fc[0].data)/np.log10(fc[0].data.max()*0.05)
alpha[np.isnan(alpha)] = 0.0
np.clip(alpha, a_min=0.0, a_max=1.0, out=alpha)
```

```python
rgb = np.transpose(np.array([redd,greend,blued]), axes=(1,2,0)).astype("float32")
```

```python
rgb *= alpha[:,:,np.newaxis]
```

```python
plt.figure(figsize=(10,10))
plt.imshow(rgb, origin='lower')
plt.axis("off")
plt.savefig("cgols_vmap.png", bbox_inches="tight", dpi=150)
```

```python
fcube = fits.open("cgols_cube_90deg.fits")
```

```python
fcube[0].data.shape
```

```python
plt.figure(figsize=(10,10))
plt.imshow(np.abs(fcube[0].data.sum(axis=0)), origin='lower', norm=LogNorm())
```

```python
cube = fcube[0].data-fcube[0].data.min(axis=0)
```

```python
f = h5py.File("spec_table.h5", "r")
vels = f["vel"][:]
f.flush()
f.close()
```

```python
range1 = (0, np.searchsorted(vels, -70.0)-1)
range2 = (range1[1], np.searchsorted(vels, 70.0)-1)
range3 = (range2[1], vels.size)
```

```python
R = cube[range3[0]:range3[1],:,:].sum(axis=0)
G = cube[range2[0]:range2[1],:,:].sum(axis=0)
B = cube[range1[0]:range1[1],:,:].sum(axis=0)
I = R+G+B
R /= I
G /= I
B /= I
R[np.isnan(R)] = 0.0
G[np.isnan(G)] = 0.0
B[np.isnan(B)] = 0.0
```

```python
Rd = convolve(R, kernel)
Bd = convolve(B, kernel)
Gd = convolve(G, kernel)
A = convolve(alpha, kernel)
```

```python
rgb = np.transpose(np.array([Rd,Gd,Bd]), axes=(1,2,0)).astype("float32")
rgb *= alpha[:,:,np.newaxis]
```

```python
plt.figure(figsize=(10,10))
plt.imshow(rgb, origin='lower')
```

```python
hdur = fits.ImageHDU(data=Rd*A)
hdur.name = "red"
hdur.writeto("red.fits", overwrite=True)
hdub = fits.ImageHDU(data=Bd*A)
hdub.name = "blue"
hdub.writeto("blue.fits", overwrite=True)
hdug = fits.ImageHDU(data=Gd*A)
hdug.name = "green"
hdug.writeto("green.fits", overwrite=True)
```

```python
fits.ImageHDU(data=rgb).writeto("rgb.fits", overwrite=True)
```
